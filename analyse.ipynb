On remarque plusieurs faits intéressants, dans l'ordre :

- La latitude est exclusivement située entre 57 et 60.

- La longitude est située environ entre 22 et 28.

- Comme énoncé dans la description des données, la valeur de hours_ahead ne dépasse pas 50 heures et est partiellement équilibrée.

- La température est majoritairement positive, avec une majorité aux alentours de 0 degrés.

- Le dewpoint est majoritairement à 0.

- Les données sur la couverture des nuages (low/mid/high) sont soit proches de 0 (ciel éclairci), soit proches de 1 (ciel couvert) avec une majorité relative de 0.

- La vitesse du vent varie mais est souvent proche de 0.

- La couverture totale de nuage est majoritairement à 1 (cumul et croisement des 3 données)

- La grande majorité des valeurs sur la radiation solaire est proche de 0.

- La quantité de neige et de précipitation sont quasi nulles.

Que peut-on en conclure ?

On peut en conclure tout d'abord que les données sont enregistrées dans une zone de 3 de latitude et 6 de longitude, avec jusqu'à 48 heures de décalage avec la base météorologique. La température est souvent proche de 0, mais il n'y a ni vent ni précipitation. Or, il y a beaucoup de nuages, avec un ciel très souvent couvert. La couverture thermique annuelle est située en moyenne entre 0 et 25 degrés. On peut en déduire qu'on se rapproche d'un climat continental russo-polonais, avec des hivers très froids et des étés chauds, mais surtout un climat très sec (peu de précipitations, de vent). On peut déjà suggérer que les données de précipitation ne seront pas pertinentes à l'analyse et l'entraînement du modèle.

```python
plt.figure(figsize=(16,16))

sns.heatmap(df_forecast.select_dtypes(['number']).corr(), annot=True)
```

On a tout d'abord les corrélations très fortes (x > 0.5 ou x < -0.5) :

- Toutes les données de couverture nuageuse (low, mid, high, total) +

- La température et le dewpoint +

Ensuite, les corrélations fortes (0.5 > x > 0.25 ou -0.25< x < 0.5) :

- Les précipitations et la neige +

- Les nuages et les précipitation / neige +

- La température et le dewpoint avec la radiation solaire +

- Les nuages et la radiation solaire -

- Les nuages et la température / le dewpoint -

On peut expliquer la forte corrélation entre les données de couverture du ciel par le fait que ce soit un cumul et que, plus il y a de nuages bas, plus il y a de chances d'avoir de nuages moyens (mais pas hauts, ceux-ci ne corrèlent qu'avec les moyens et les totaux)

Concernant les données de température, elles corrèlent avec le dewpoint car plus le soleil est proche du zénith, plus il rayonne et émet de chaleur (moment de la journée le plus chaud en général). Elles corrèlent aussi avec les données de radiation solaire comme expliqué précédemment.

La neige étant une précipitation, elle est comptabilisée dans le calcul d'une précipitation totale et c'est la raison pour laquelle une forte corrélation existe. De même pour la présence de nuages avec ces deux données car ce sont ces mêmes nuages qui causent les précipitations / neiges.

Enfin, les nuages obstruant la radiation solaire, ils corrèlent avec cette donnée mais également la température et le dewpoint.

```python
sns.scatterplot(data=df_forecast, x='temperature', y='dewpoint')
```

On peut voir une nette augmentation de la température en fonction du dewpoint.

```python
sns.scatterplot(data=df_forecast, x='snowfall', y='total_precipitation')
```

Malgré une très grande majorité de valeurs nulles, on peut remarquer une croissance de la précipitation totale en fonction de la neige.

```python
sns.scatterplot(data=df_forecast, x='cloudcover_low', y='cloudcover_total')
```

On remarque donc la forte corrélation entre la couverture du ciel par les nuages bas et la couverture totale du ciel.

Cependant, il serait intéressant de les analyser couplées aux nouvelles données récupérées sur ce dataset (windspeed, winddirection, shortwave et surface_pressure (on ne prend pas rain en compte car ce n'est que total_precipitation soustrait de snowfall)).

Nous remarquons donc : 

- La vitesse du vent en majorité entre 0 et 15 km/h

- La direction du vent majoritairement orientée vers 200 degrés.

- La shortwave radiation majoritairement à 0 mais allant jusqu'à 800 Wh

- La pressure à la surface majoritairement à 1 010 hectopascals

- La radiation diffuse du soleil majoritairement à 0 mais avec une part conséquente située entre 0 et 100 Wh.

```python
plt.figure(figsize=(16,16))

sns.heatmap(data=df_weather.select_dtypes(['number']).corr(), annot=True)
```

Dans les nouvelles corrélations très fortes, nous avons : 

- Shortwave_radiation, direct_solar_radiation et diffuse_radiation +

Concernant les nouvelles corrélations fortes, il y a :

- Les radiations avec la température et le dewpoint +

- Les radiations avec la couverture du ciel -

- La longitude et la vitesse du vent -

- La pression à la surface et la couverture du ciel -

- La pression à la surface avec les précipitations (pluie, neige) -

Les radiations étant intrinsèquement liées (différents types de radiations venant d'une même source, le soleil), plus une radiation est élevée plus les autres le seront (leur ratio est presque fixe).

Les radiations étant plus détaillées, elles gardent le même effet sur le zénith et la température. Plus le soleil s'approche du zénith, plus il émet de radiations, plus la température est élevée.

De même pour la couverture du ciel, plus le ciel est couvert, moins les radiations passent.

Concernant la corrélation entre la couverture du ciel et la pression à la surface, elle s'explique par le fait que les nuages sont formés par une pression de masse d'air qui est refroidie par la prise d'altitude. S'il y a des nuages, la pression atmosphérique a été convertie et diminue donc.

Je n'ai malheureusement pas trouvé d'explication à la corrélation entre la longitude et la vitesse du vent, à moins d'une corrélation avec la pression atmosphérique étant plus forte dans une région à basse longitude et générant alors un vent plus fort.

```python
sns.scatterplot(data=df_weather, y='longitude', x='windspeed_10m')
```

On peut voir que la vitesse du vent est généralement plus élevée dans les basses longitudes, comme conjecturé précédemment.
